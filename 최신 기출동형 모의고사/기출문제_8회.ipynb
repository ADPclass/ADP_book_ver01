## 머신러닝 -  은행 고객 이탈 예측 (Exited가 종속변수) 

### EDA 및 전처리: 데이터의 특징을 파악하고 분석에 적합하도록 전처리를 수행하시오.

```python
import pandas as pd
import numpy as np

# 데이터 로드
url = "https://raw.githubusercontent.com/sharmaroshan/Churn-Modelling-Dataset/master/Churn_Modelling.csv"
df = pd.read_csv(url)
df
```

① 결측치 확인 및 처리
데이터의 완전성을 확인하기 위해 결측치를 체크합니다. 본 데이터셋에는 결측치가 없으나, 만약 존재한다면 수치형은 평균/중앙값으로, 범주형은 최빈값으로 대체하거나 삭제하는 전략을 취합니다.

```python
df = df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)

# 결측치 확인
print(df.isnull().sum())
```

② 이상치(Outlier) 탐색 및 처리 (수정)
이상치는 모델의 성능을 왜곡할 수 있으므로 주의 깊게 다뤄야 합니다. 무작정 제거하기보다, 먼저 시각화를 통해 이상치의 분포를 확인하고 비즈니스적 관점에서 처리 여부를 결정해야 합니다.

Age 변수의 분포를 Boxplot으로 그려 이상치가 어느 정도 존재하고 있는지 확인합니다.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. 시각화 공간 설정 (2행 3열 구성)
plt.figure(figsize=(15, 10))

# 2. 각 변수별 박스플롯을 하나씩 배치
plt.subplot(2, 3, 1)
sns.boxplot(y=df['CreditScore'])
plt.title('CreditScore')

plt.subplot(2, 3, 2)
sns.boxplot(y=df['Age'])
plt.title('Age')

plt.subplot(2, 3, 3)
sns.boxplot(y=df['Tenure'])
plt.title('Tenure')

plt.subplot(2, 3, 4)
sns.boxplot(y=df['Balance'])
plt.title('Balance')

plt.subplot(2, 3, 5)
sns.boxplot(y=df['NumOfProducts'])
plt.title('NumOfProducts')

plt.subplot(2, 3, 6)
sns.boxplot(y=df['EstimatedSalary'])
plt.title('EstimatedSalary')

# 3. 레이아웃 조정 및 출력
plt.tight_layout()
plt.show()
```

```python
#CreditScore, Age, NumOfProducts 에 이상치 존재 

# 1. 수치형 컬럼만 자동으로 선택 (ID, 범주형 제외)
# select_dtypes를 쓰면 수치형만 싹 골라낼 수 있습니다.
numeric_df = df.select_dtypes(include=[np.number]).drop(columns=['Exited'], errors='ignore')

# 2. 전 컬럼의 Q1, Q3, IQR을 동시에 계산
# 리스트를 일일이 만들 필요 없이 표 전체에서 계산합니다.
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

# 3. 이상치 범위 설정 (표 전체에 대한 경계값)
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 4. [핵심] 이상치가 하나라도 있는 행(Row) 찾아내기
# 각 행마다 모든 컬럼을 검사해서 범위를 벗어나는지 체크합니다.
is_outlier = (numeric_df < lower_bound) | (numeric_df > upper_bound)

# 5. 이상치가 없는 깨끗한 행만 남기기
# any(axis=1)은 '가로 방향으로 하나라도 이상치가 있으면 True'라는 뜻입니다.
# ~ 기호를 붙여서 '이상치가 하나도 없는 행'만 골라냅니다.
df_clean = df[~is_outlier.any(axis=1)]

print(f"제거 전 데이터: {len(df)}개")
print(f"제거 후 데이터: {len(df_clean)}개")
```

③ 범주형 변수 원-핫 인코딩(One-Hot Encoding)
머신러닝 알고리즘은 문자열을 직접 이해할 수 없으므로, Geography와 Gender 같은 범주형 변수를 수치형인 더미 변수로 변환합니다. 이때 다중공선성을 방지하기 위해 drop_first=True 옵션을 사용합니다.

```python
#  범주형 변수 인코딩 (이상치가 제거된 데이터 사용)
df_final = pd.get_dummies(df_clean, columns=['Geography', 'Gender'], drop_first=True)
```

④ 데이터 스케일링(Scaling) (추가)
이제 전처리의 마지막 단계인 스케일링을 진행합니다.

```python
#  스케일링 (전체 수치 데이터의 단위를 맞춤)
from sklearn.preprocessing import StandardScaler

X = df_final.drop('Exited', axis=1)
y = df_final['Exited']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 학습을 위해 데이터프레임 형태로 유지 (선택사항)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled_df
```

② 데이터 분할 (Train / Test Split)
주의: 오버샘플링을 하기 전에 반드시 데이터를 먼저 나눠야 합니다. 테스트 데이터까지 불려버리면 데이터 누수(Data Leakage)가 발생할 수 있기 때문입니다.

```python
from sklearn.model_selection import train_test_split

# 독립변수와 종속변수 설정 (이미 전처리된 X_scaled 사용)
# 여기서는 작가님 원고 흐름에 맞춰 X_scaled와 y를 사용합니다.
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, stratify=y, random_state=42)

print(f"학습 데이터 크기: {X_train.shape}")
print(f"테스트 데이터 크기: {X_test.shape}")
```

### 모델링: 로지스틱 회귀를 사용하여 모델을 생성하고, 변수의 유의성을 검정 및 해석하시오.

```python
import statsmodels.api as sm

# 1. 로지스틱 회귀 모델(Logit) 생성 및 학습
logit_model = sm.Logit(y_train_res, X_train_const).fit()

# 2. 모델 결과 요약 출력
print(logit_model.summary())
```

## 통계분석(40점)

### 통계적 가설 검정 (10점)
3가지 서로 다른 교육 방법(그룹 A, B, C)에 따른 math score 차이가 있는지 검증하고자 한다. 수집된 데이터를 바탕으로 일원분산분석(One-way ANOVA)을 수행하고, 결과에 따른 분석을 실시하시오.

```python
import pandas as pd
import numpy as np
# 1. 데이터 로드 및 준비
url = 'https://raw.githubusercontent.com/ADPclass/ADP_book_ver01/main/data/students_performance_manipulated.csv'
df = pd.read_csv(url)
df
```

```python
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# 분석을 위해 'group A', 'group B', 'group C'만 추출 (성적은 'math score' 기준)
df_anova = df[df['race/ethnicity'].isin(['group A', 'group B', 'group C'])]

# 2. 가설 설정
# 귀무가설(H0): 세 교육 방법에 따른 성적 평균은 모두 같다.
# 대립가설(H1): 적어도 한 교육 방법의 성적 평균은 서로 다르다.

# 3. ANOVA 수행
group_a = df_anova[df_anova['race/ethnicity'] == 'group A']['math score']
group_b = df_anova[df_anova['race/ethnicity'] == 'group B']['math score']
group_c = df_anova[df_anova['race/ethnicity'] == 'group C']['math score']

f_stat, p_val = stats.f_oneway(group_a, group_b, group_c)

print(f"ANOVA 결과: F-통계량 = {f_stat:.4f}, p-value = {p_val:.4f}")

# 4. 사후 분석 (Tukey HSD)
if p_val < 0.05:
    posthoc = pairwise_tukeyhsd(df_anova['math score'], df_anova['race/ethnicity'], alpha=0.05)
    print(posthoc)
```

[ANOVA 결과]

F-통계량: 1145.02

p-value: 2.30e-90 (0.05보다 매우 작음)

판정: 귀무가설을 기각하며, 교육 방법 간 성적 평균 차이가 통계적으로 매우 유의미하다. 

[사후 분석(Tukey HSD) 결과]

group A vs group B: 유의미한 차이 있음 (A가 약 15점 높음)

group A vs group C: 유의미한 차이 있음 (C가 약 30점 높음)

group B vs group C: 유의미한 차이 있음 (C가 약 45점 높음)

결론: 통계적으로 모든 그룹 간의 성적 차이가 유의미하며, 교육 방법 C가 가장 효과적인 것으로 나타났다.

### 시계열 분석 (30점)
데이터셋: 월별 판매량 데이터 (sales_data.csv)

```python
import pandas as pd

# 1. 데이터 로드 (월별 자동차 판매량, 108개월치(과거9년),미래 1년 예측 )
url = 'https://raw.githubusercontent.com/ADPclass/ADP_book_ver01/main/data/monthly-car-sales.csv'
df_sales = pd.read_csv(url)
df_sales
```

#### ADF(Augmented Dickey-Fuller) 검정을 통해 해당 시계열 데이터의 정상성을 확인하시오.

시계열 분석의 첫 단추는 데이터가 정상성(Stationarity)을 만족하는지 확인하는 것입니다. 

정상성을 만족하지 않는 데이터(추세나 계절성이 있는 데이터)는 ARIMA 모델을 바로 적용할 수 없기 때문입니다.

```python
from statsmodels.tsa.stattools import adfuller
# ADF 검정 수행
def adf_test(series):
    result = adfuller(series)
    print(f'ADF Statistic: {result[0]:.4f}')
    print(f'p-value: {result[1]:.4f}')
    if result[1] <= 0.05:
        print("결과: 유의수준 0.05에서 정상성을 만족함 (H0 기각)")
    else:
        print("결과: 유의수준 0.05에서 정상성을 만족하지 못함 (H0 채택)")

print("--- 원본 데이터 ADF 검정 ---")
adf_test(df_sales['Sales'])
```

```python
# 1차 차분 수행
df_diff = df_sales['Sales'].diff().dropna()

# 차분된 데이터로 다시 ADF 검정 수행
print("--- 1차 차분 후 ADF 검정 ---")
result_diff = adfuller(df_diff)
print(f'ADF Statistic: {result_diff[0]:.4f}')
print(f'p-value: {result_diff[1]:.4f}')

if result_diff[1] <= 0.05:
    print("결론: 1차 차분 후 데이터가 정상성을 만족함.")
else:
    print("결론: 여전히 비정상성임. 추가 차분이 필요할 수 있음.")
```

#### ARIMA 모델을 구축하되, 최적의 차수(p, d, q)를 결정하는 과정을 상세히 기술하시오.

"원본 데이터에 대한 ADF 검정 결과, p-value가 0.05보다 크게 나타나 귀무가설(H_0: 시계열에 단위근이 존재하며 정상적이지 않다)을 채택합니다.

즉, 본 데이터는 추세가 존재하는 비정상 시계열이므로 ARIMA 모델 구축을 위해 차분(Differencing) 과정이 필요합니다."

```python
from pmdarima import auto_arima

# 최적의 차수 탐색 (계절성이 뚜렷하므로 seasonal=True 설정)
# m=12는 월별 데이터의 1년 주기를 의미합니다.
SARIMA_model = auto_arima(df_sales['Sales'], 
                           start_p=1, start_q=1,
                           max_p=3, max_q=3, 
                           m=12, start_P=0, 
                           seasonal=True, d=1, D=1, 
                           trace=True, error_action='ignore',  
                           suppress_warnings=True, stepwise=True)

print(SARIMA_model.summary())
```

ARIMA 모델 구축 및 최적 차수($p, d, q$) 결정 결과auto_arima를 통해 수천 개의 조합을 탐색한 결과, AIC(Akaike Information Criterion)를 최소화하는 최적의 모델로 SARIMA(1, 1, 1)(0, 1, 1)_{12}가 선정되었습니다.

① 모델 파라미터 결정 근거

AIC 점수: 1683.747 (가장 낮은 수치로, 모델의 적합도와 간결성이 가장 균형 잡힌 상태)

차수($p, d, q$): 1차 차분($d=1$)을 통해 추세를 제거하고, 1차 자기회귀($p=1$)와 1차 이동평균($q=1$) 모델을 결합함.

계절성($P, D, Q, s$): 월별 데이터의 12개월 주기를 반영하기 위해 계절 차분($D=1$)과 계절 이동평균($Q=1$)을 적용한 Seasonal ARIMA 모델이 최종 선택됨.

#### 구축된 모델을 활용하여 향후 12개월의 판매량을 예측하고 이를 시각화하시오.

```python
import matplotlib.pyplot as plt
import pandas as pd

# [핵심 교정] 시각화 직전에 원본 데이터의 인덱스를 날짜로 강제 전환합니다.
# 만약 'Month' 컬럼이 보이지 않는다면 df_sales.index = pd.to_datetime(df_sales.index)를 시도하세요.
if 'Month' in df_sales.columns:
    df_sales['Month'] = pd.to_datetime(df_sales['Month'])
    df_sales.set_index('Month', inplace=True)
else:
    df_sales.index = pd.to_datetime(df_sales.index)

# 1. 향후 12개월 예측 및 신뢰구간 산출
forecast, conf_int = SARIMA_model.predict(n_periods=12, return_conf_int=True)

# 2. 예측값 전용 날짜 인덱스 생성
# freq='MS'는 월초(1일) 기준입니다. 데이터에 따라 'M'으로 바꿀 수 있습니다.
forecast_index = pd.date_range(df_sales.index[-1], periods=13, freq='MS')[1:]

# 3. 그래프 시각화
plt.figure(figsize=(12, 6))

# 원본 데이터 플롯 (인덱스가 날짜여야 에러가 나지 않습니다)
plt.plot(df_sales.index, df_sales['Sales'], label='Actual Sales', color='blue')

# 예측 데이터 플롯
plt.plot(forecast_index, forecast, label='Forecast (12 Months)', color='red', linestyle='--')

# 신뢰구간 표시
plt.fill_between(forecast_index, 
                 conf_int[:, 0], conf_int[:, 1], 
                 color='pink', alpha=0.3, label='95% Confidence Interval')

plt.title('Monthly Car Sales: 12-Month Forecast (35th ADP)')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True, alpha=0.3)

# 4. 최종 출력
plt.show()
```
